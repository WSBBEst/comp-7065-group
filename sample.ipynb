# Polyvore Route A (缺图过滤) Demo

运行方式：依次执行下面的 cell。

```python
from pathlib import Path

def find_project_root() -> Path:
    here = Path.cwd().resolve()
    for p in [here, *here.parents]:
        if (p / "polyvore_route_a.py").exists():
            return p
    raise FileNotFoundError("polyvore_route_a.py not found in current directory or parents")

project_root = find_project_root()
project_root
```

```python
import os

images_root = project_root / "archive" / "Re-PolyVore" / "Re-PolyVore" / "Re-PolyVore"
polyvore_tar = project_root / "polyvore-dataset-master" / "polyvore.tar.gz"

images_root.exists(), polyvore_tar.exists(), str(images_root), str(polyvore_tar)
```

```python
import subprocess
import sys

cmd = [
    sys.executable,
    str(project_root / "polyvore_route_a.py"),
    "stats",
    "--images-root",
    str(images_root),
]
result = subprocess.run(cmd, check=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
result
```

```python
import subprocess
import sys

cmd = [
    sys.executable,
    str(project_root / "polyvore_route_a.py"),
    "train-cp",
    "--images-root",
    str(images_root),
    "--epochs",
    "5",
    "--batch-size",
    "8",
    "--arch",
    "bilstm",
    "--lstm-hidden",
    "128",
    "--freeze-backbone",
]
result = subprocess.run(cmd, check=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
result
```

```python
import subprocess
import sys

fitb_ckpt = project_root / "fitb_ckpt_demo.pt"

cmd = [
    sys.executable,
    str(project_root / "polyvore_route_a.py"),
    "train-fitb",
    "--images-root",
    str(images_root),
    "--train-samples",
    "5000",
    "--epochs",
    "2",
    "--batch-size",
    "4",
    "--freeze-backbone",
    "--save-path",
    str(fitb_ckpt),
]
result = subprocess.run(cmd, check=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
result
```

```python
import subprocess
import sys

cmd = [
    sys.executable,
    str(project_root / "polyvore_route_a.py"),
    "eval-fitb",
    "--images-root",
    str(images_root),
    "--checkpoint",
    str(fitb_ckpt),
    "--batch-size",
    "4",
]
result = subprocess.run(cmd, check=True, capture_output=True, text=True)
print(result.stdout)
print(result.stderr)
result
```
